```python
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

```python
def DF_to_MONDB(df, collection):
    for i in range(df.shape[0]):
        vac = df.loc[i]
        dictionary = {
            'name': vac['name'],
            'link': vac["link"],
            'mesto': vac["mesto"],
            'org':vac["org"],
            'min': vac["min"],
            'max': vac['max'],
            'valuta': vac["valuta"],
            'resurse': vac["resurse"]}
        collection.insert_one(dictionary)
```

```python
# подключаемся к mongodb
client = MongoClient('localhost',27017)
db = client['vakancy']
```
